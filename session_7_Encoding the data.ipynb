```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

visa_df=pd.read_csv(r'C:\Users\rajat\OneDrive\Desktop\DS GenAI\Data Files\Visadataset.csv')
cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

**Map**

```python
visa_df['case_status'].unique()
```

```python
dict1={'Certified':0,'Denied':1}
dict1
```

```python
visa_df['case_status'].map(dict1)
```

```python
visa_df['case_status_map']=visa_df['case_status'].map(dict1)
visa_df
```

```python
visa_df['continent'].unique()
```

```python
labels=visa_df['continent'].unique()
d={labels[i]:i for i in range(len(labels))}
visa_df['continent']=visa_df['continent'].map(d)
visa_df
```

```python
for i in cat[1:]:
    labels=visa_df[i].unique()
    d={labels[i]:i for i in range(len(labels))}
    visa_df[i]=visa_df[i].map(d)
visa_df
```
