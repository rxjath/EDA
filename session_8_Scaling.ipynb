```python
#Step-1: take prevailing wage data
#Step-2: calculate mean : u
#Step-3: calculate SD : sigma
#Step-4: Nr=step1-step2
#Step-5: Z=Nr/step3
#Step-6: Create a new column prevailingwage _z
#Step-7: Draw the distribution plot for both prevailing wage and prevailingwage_z

#Observation : Even though values are modified data 
```

```python
import warnings
warnings.filterwarnings('ignore')
wage_data=visa_df['prevailing_wage']
u=wage_data.mean()
SD=wage_data.std()
Nr=wage_data-u
Z=Nr/SD

visa_df['prevailing_wage_Z']=Z

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(visa_df['prevailing_wage'],bins=40)
plt.subplot(1,2,2)
sns.histplot(visa_df['prevailing_wage_Z'],bins=40)
plt.show()

```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1).hist(visa_df['prevailing_wage'],bins=40)
plt.subplot(1,2,2).hist(visa_df['prevailing_wage_Z'],bins=40)
plt.show()
```
