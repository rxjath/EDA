Seperate categorical and numerical columns

#idea

make two empty list cat_list,num_list

iterate the data_types.items or zip(keys,values)

if value==int or float then key should got to numlist

```python
cat_list,num_list=[],[]
for key,value in zip(keys,values):
    if value=='int64':
        num_list.append(key)
    else:
        cat_list.append(key)

print('Numerical :',num_list)
print('Categorical :',cat_list)
```

```python
dic=dict(data_types)
dic
```

```python
cat_list,num_list=[],[]
for key,value in dic.items():
    if value=='object':
        num_list.append(key)
    else:
        cat_list.append(key)

print('Numerical :',num_list)
print('Categorical :',cat_list)
```

```python
num_list = [key for key, value in dic.items() if value != 'object']
cat_list = [key for key, value in dic.items() if value == 'object']

print('Numerical:', num_list)
print('Categorical:', cat_list)
```
